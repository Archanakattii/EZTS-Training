```python
#Double hashing

l=[20,34,45,70,56,81,104,37,46,39]
hash_t=[False]*11
for i in l:
    h1_k=i%11
    for j in range(len(l)):
        h2_k=8-(i%8)
        h_t=(h1_k+j*(h2_k))%11
        if hash_t[h_t]==False:
            hash_t[h_t]=i
            break
print(hash_t)
```

```python
#dynamic programming

def dp(l,b,n):
    if n==0 or b==0:
        return 0
    if (l[n-1]>b):
        return dp(l,b,n-1)
    else:
        return max(l[n-1]+dp(l,b-l[n-1],n-1),dp(l,b,n-1))

l=[1,5,7]
b=18
n=len(l)
print(dp(l,b,n))
```
