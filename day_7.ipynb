```python
#TREE
#DFS-preorder,inorder,postorder

class node:
    def __init__(self,data):
        self.value=data
        self.left=None
        self.right=None
    
def preorder(root):           #Root,left,right
    if root == None:
        return
    print(root.value)
    preorder(root.left)
    preorder(root.right)
    
    
def inorder(root):             #left,Root,right
    if root==None:
        return
    inorder(root.left)
    print(root.value,end=",")
    inorder(root.right)
    
    
def postorder(root):          #left,right,Root
    if root==None:
        return
    postorder(root.left)
    postorder(root.right)
    print(root.value,end=" ")


if __name__=="__main__":
    root=node(1)
    
    root.left=node(2)
    root.right=node(3)
    
    root.left.left=node(4)
    root.left.right=node(5)
    
    root.right.left=node(6)
    root.right.right=node(7)
    
    preorder(root)
    inorder(root)
    postorder(root)
    
```

```python
#BFS
#level order

class node:
    def __init__(self,data):
        self.value=data
        self.left=None
        self.right=None
        
def levelorder(root):
    Q=[root]
    Q.append(None)
    
    while len(Q)>0:
        curr = Q.pop(0)
        
        if curr == None:
            if len(Q)==0:
                break
            else:
                print()
                Q.append(None)
        
        else:
            print(curr.value,end=" ")
            if curr.left != None:
                Q.append(curr.left)
            if curr.right != None:
                Q.append(curr.right)
                          
        
if __name__=="__main__":
    root=node(1)
    
    root.left=node(2)
    root.right=node(3)
    
    root.left.left=node(4)
    root.left.right=node(5)
    
    root.right.left=node(6)
    root.right.right=node(7)
    
    root.left.right.left=node(9)
    root.left.right.right=node(10)
    root.right.right.right=node(11)
    
    root.left.right.left.left=node(12)
    root.left.right.left.right=node(13)
    
    
    levelorder(root)
    
```

```python
#height of the tree

class node:
    def __init__(self,data):
        self.value=data
        self.left=None
        self.right=None
        
def height(root):
    if root==None:
        return 0
    LH=height(root.left)
    RH=height(root.right)
    return max(LH,RH)+1
           
        
if __name__=="__main__":
    root=node(1)
    
    root.left=node(2)
    root.right=node(3)
    
    root.left.left=node(4)
    root.left.right=node(5)
    
    root.right.left=node(6)
    root.right.right=node(7)
    
    root.left.right.left=node(9)
    root.left.right.right=node(10)
    root.right.right.right=node(11)
    
    root.left.right.left.left=node(12)
    root.left.right.left.right=node(13)
    
    print(height(root))
    
```

```python
#to print the leaf of the tree

class node:
    def __init__(self,data):
        self.value=data
        self.left=None
        self.right=None
        
        
def leaf(root):
    if not root:
        return None
    
    Q=[root]
    l=None
    
    while len(Q)>0:
        curr = Q.pop(0)
        
        if (curr.left and curr.right)==None:
            l=curr.value
        if curr.left!=None:
            Q.append(curr.left)
        if curr.right!=None:
            Q.append(curr.right)
    return l

if __name__=="__main__":
    root=node(1)
    
    root.left=node(2)
    root.right=node(3)
    
    root.left.left=node(4)
    root.left.right=node(5)
    
    root.right.left=node(6)
    root.right.right=node(7)
    
    root.left.right.left=node(9)
    root.left.right.right=node(10)
    root.right.right.right=node(11)
    
    root.left.right.left.left=node(12)
    root.left.right.left.right=node(13)
    
    #root.left.right.left.right.right=node(14)
    
    print(leaf(root))
```

```python
#TOP VIEW

class node_data:
    def __init__(self,Node,Hkey):
        self.node=Node
        self.hkey=Hkey
        
def topview(root):
    temp=node_data(root,0)
    Q=[temp]
    Q.append(None)
    
    key_dict={}
    
    while len(Q)!=0:
        curr=Q.pop(0)
        
        
        if curr==None:
            if len(Q)==0:
                break
            else:
                Q.append(None)
        else:
            if curr.hkey not in key_dict.keys():
                key_dict[curr.hkey]=curr.node.value
                
            if curr.node.left!=None:
                temp=node_data(curr.node.left,curr.hkey-1)
                Q.append(temp)
                
            if curr.node.right!=None:
                temp=node_data(curr.node.right,curr.hkey+1)
                Q.append(temp)
    
    for i in sorted(key_dict):
        print(key_dict[i])
        
    print(key_dict)
    

    
topview(root)
            
                
```
